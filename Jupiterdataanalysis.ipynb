**removing outliers using interquartile range(IQR),upper bound and lower bound of each column** 

here you find lower bound and upper bound of each column then we replace the number in the column if it is greater than upper bound or less than lower bound with mean() of the column

```python
import numpy as np
data5=data4.copy()

#Function to replace outliers with the mean of the column using IQR
def replace_outliers_with_mean_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        df[column] = np.where(df[column] > upper_bound, df[column].mean(), df[column])
        df[column] = np.where(df[column] < lower_bound, df[column].mean(), df[column])
        
    return df

# List of columns to process
columns_to_process =columns_to_plot
data5 = replace_outliers_with_mean_iqr(data5, columns_to_process)
```

# creating two  new variables 
1. **Total_Alcohol_Consumption: Sum of Workday and Weekend Alcohol Consumption (Workday_Alcohol_Consumption + Weekend_Alcohol_Consumption)**

2. **Average_Grade: Average of the first and second semester grades ((First_Semester_Grade + Second_Semester_Grade ) / 2)**

3. **Study_Efficiency: Ratio of study time to travel time (studytime / traveltime)**

```python
# creating copy
data8=data7.copy()


#creating new columns
data8['Total_Alcohol_Consumption'] = data8['Workday_Alcohol_Consumption'] + data8['Weekend_Alcohol_Consumption']
data8['Average_Grade'] = (data8['First_Semester_Grade'] + data8['Second_Semester_Grade']) / 2
data8['Study_Efficiency'] = data8['Study_Time'] / data8['Travel_Time']


# Viewing dataset with new added columns
print(data8.head())

```

# removing outliers of those new introduced columns

```python
#creating copy
data9=data8.copy()

# Function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers for the new variables
new_columns = ['Total_Alcohol_Consumption', 'Average_Grade', 'Study_Efficiency']

for column in new_columns:
    data9= remove_outliers(data8, column)

# Display the cleaned DataFrame
print(data9.head())
```

# encoding categorical variable

**changing column named 'Mother_Job', 'Father_Job', 'Reason', 'Guardian', 'School_Support' values into numbers**

```python
# importing package used for encoding
from sklearn.preprocessing import LabelEncoder


#creating copy
data10=data9.copy()

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Columns to encode
columns_to_encode = ["Mother_Job", "Father_Job", "Reason", "Guardian", "School_Support"]

# Encode the specified columns among all columns
for column in columns_to_encode:
    data10[column] = label_encoder.fit_transform(data10[column])

# Display the DataFrame to verify the changes
print(data10.head())
```
